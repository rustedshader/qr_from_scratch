```python
from PIL import Image
im = Image.new('L',(25,25))


def draw_alignment_area(x,y):
    for i in range(7):
        for k in range(7):
            im.putpixel((x+i,y+k),255)
    for i in range(2,5):
        for k in range(2,5):
            im.putpixel((x+i,y+k),0)

draw_alignment_area(0,0)
draw_alignment_area(0,18)
draw_alignment_area(18,0)
zoomed = im.resize((250, 250), Image.NEAREST)
display(zoomed)
```

```python
def draw_encoding():
    byte = [0,1,0,0]
    x = 24
    y = 24
    for bt,index in enumerate(byte):
        if bt == 0: 
            if index % 2 == 0:
                im.putpixel((x-index,y),255)
            else:
                im.putpixel((x,y-index),255)


draw_encoding()
zoomed = im.resize((250, 250), Image.NEAREST)
display(zoomed)
```
